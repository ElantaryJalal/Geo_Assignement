## Correlation and Data analysis

From experience, we expect `porosity` and `permeability` to be correlated. This can be checked by calculating the correlation coefficient between two parameters. 

$$\rho_{x,y}=\frac{Cov_{x,y}}{\sigma_x \sigma_y}$$

with

$$Cov_{x,y}=\frac{\sum_{i=1}^{n}(x_{i}-\bar{x})(y_{i}-\bar{y})}{n}$$

<br>

<div class="alert alert-block alert-info">
Your task: Complete the function to calculate the corellation coefficient and round it to 4 decimal points. 
<div>

````python
def calc_cor_coeff(x, y):
    """Compute the correlation coefficient between two arrays x and y """
    
    # YOUR CODE HERE
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    n = len(x)
    cov = np.sum((x - x_mean) * (y - y_mean)) / n
    sigma_x = np.std(x)
    sigma_y = np.std(y)
    correlation = np.round((cov / (sigma_x * sigma_y)),4)
    return correlation

porosity = data['porosity']
permeability = data['permeability']
print(calc_cor_coeff(porosity,permeability))
    
````

````python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
#data = pd.read_csv("assignment_dataset.csv")

# --- Manual correlation function (according to the given formula) ---
def calc_corr(x, y):
    """
    Compute Pearson correlation using:
      rho = Cov(x,y) / (sigma_x * sigma_y)
    with population covariance/STD (denominator n), rounded to 4 decimals.
    Drops NaNs pairwise.
    """
    xy = pd.DataFrame({"x": x, "y": y}).dropna()
    x_vals = xy["x"].to_numpy()
    y_vals = xy["y"].to_numpy()

    x_mean = np.mean(x_vals)
    y_mean = np.mean(y_vals)

    cov_xy  = np.sum((x_vals - x_mean) * (y_vals - y_mean)) / len(x_vals)
    sigma_x = np.std(x_vals)  # population std (ddof=0)
    sigma_y = np.std(y_vals)

    rho = cov_xy / (sigma_x * sigma_y)
    return np.round(rho, 4)

# --- Calculate correlation coefficients ---
corr_phi_k  = calc_corr(data['porosity'],            data['permeability'])
corr_phi_ai = calc_corr(data['porosity'],            data['acoustic_impedance'])
corr_k_ai   = calc_corr(data['permeability'],        data['acoustic_impedance'])

print("Correlation (Porosity vs Permeability):", corr_phi_k)
print("Correlation (Porosity vs Acoustic Impedance):", corr_phi_ai)
print("Correlation (Permeability vs Acoustic Impedance):", corr_k_ai)

# --- Create scatter plot subplots ---
fig, axs = plt.subplots(1, 3, figsize=(21, 7), constrained_layout=True)

# 1) Porosity vs Permeability
axs[0].scatter(data['porosity'], data['permeability'], s=25, alpha=0.8,
               edgecolor='black', linewidths=0.5)
axs[0].set_title(f"Porosity vs Permeability\nÏ = {corr_phi_k}")
axs[0].set_xlabel("Porosity (%)")
axs[0].set_ylabel("Permeability (mD)")
axs[0].grid(True, linestyle='--', alpha=0.4)

# 2) Porosity vs Acoustic Impedance
axs[1].scatter(data['porosity'], data['acoustic_impedance'], s=25, alpha=0.8,
               edgecolor='black', linewidths=0.5)
axs[1].set_title(f"Porosity vs Acoustic Impedance\nÏ = {corr_phi_ai}")
axs[1].set_xlabel("Porosity (%)")
axs[1].set_ylabel("Acoustic Impedance (PaÂ·s/m)")
axs[1].grid(True, linestyle='--', alpha=0.4)

# 3) Permeability vs Acoustic Impedance
axs[2].scatter(data['permeability'], data['acoustic_impedance'], s=25, alpha=0.8,
               edgecolor='black', linewidths=0.5)
axs[2].set_title(f"Permeability vs Acoustic Impedance\nÏ = {corr_k_ai}")
axs[2].set_xlabel("Permeability (mD)")
axs[2].set_ylabel("Acoustic Impedance (PaÂ·s/m)")
axs[2].grid(True, linestyle='--', alpha=0.4)

plt.show()
````

For more information about dictionaries, look [here](https://docs.python.org/3/tutorial/datastructures.html#dictionaries)

````python
import pandas as pd
import numpy as np

# Load the dataset
#data = pd.read_csv("assignment_dataset (2).csv")

def find_subsets(data):
    """
    Find the different subsets in a dataset (by facies_ID) and return each subset as a dataframe.
    Returns a dictionary: {'faciesX': <subset_df>, ...}
    """
    # Get unique facies IDs (drop NaNs, sort for stable ordering)
    unique_facies = np.sort(pd.unique(data['facies_ID'].dropna()))
    
    facies_dict = {}
    for fid in unique_facies:
        # Build a clean key like 'facies1' instead of 'facies1.0'
        fid_str = str(fid)
        if fid_str.endswith('.0'):
            fid_str = fid_str[:-2]
        key = f"facies{fid_str}"
        facies_dict[key] = data[data['facies_ID'] == fid].copy()
    return facies_dict

# Find subsets
subsets = find_subsets(data)

# Print the number of unique facies IDs and a brief summary of each subset
print(f"Number of unique facies IDs: {len(subsets)}")
for facies_key, subset in subsets.items():
    print(f"{facies_key}: {len(subset)} rows")
````

````python
# YOUR CODE HERE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

def calc_cloud(data, parameter='porosity'):
    """
    calculate the distance matrix and the semivariance matrix of the input dataframe
    """
    df = data[['X', 'Y', parameter]].dropna().copy()

    coords = df[['X', 'Y']].to_numpy(float)
    z = df[parameter].to_numpy(float).reshape(-1,1)

    dist_mat = squareform(pdist(coords, metric='euclidean'))
    dz = squareform(pdist(z, metric='euclidean'))
    semivars = 0.5 * (dz**2)

    return dist_mat, semivars
````

````python
def calc_variogram(dm_flat, sv_flat, start, bin_size):
    """
    Calculate the experimental variogram
    dm_flat : np.array = flattended distance matrix (with np.flatten())
    sv_flat : np.array = flattended semivariance matrix (with np.flatten())
    start : int = starting lag
    bin_size : int = size of the bins
    
    :return: 
    
    h_bins : np.array = centers of the bins
    ave_vals : np.array = corresponding average semivariances
    """
    # YOUR CODE HERE
    dm_flat = np.asarray(dm_flat, dtype=float)
    sv_flat = np.asarray(sv_flat, dtype=float)

    # keep only finite values and exclude zero-distance pairs
    m = np.isfinite(dm_flat) & np.isfinite(sv_flat) & (dm_flat > 0)
    h = dm_flat[m]
    g = sv_flat[m]

    if h.size == 0:
        return np.array([]), np.array([])

    # choose max lag as stop, per instruction
    stop = h.max()

    # bin edges and centers
    edges = np.arange(start, stop + bin_size, bin_size)
    if edges.size < 2:  # ensure at least one bin
        edges = np.array([start, start + bin_size])
    h_bins = 0.5 * (edges[:-1] + edges[1:])

    # average semivariance in each bin
    ave_vals = np.full(h_bins.shape, np.nan, dtype=float)
    for i in range(len(h_bins)):
        sel = (h >= edges[i]) & (h < edges[i + 1])
        if np.any(sel):
            ave_vals[i] = np.mean(g[sel])

    return h_bins, ave_vals
    raise NotImplementedError()
    return h_bins, ave_vals
````

## Ordinary Kriging

With the variogram analysis, you now have everything to perform a kriging exercise on the dataset, i.e. create a kriged parameter map (what we originally wanted). 
In the exercise, we performed simple kriging on a 1D well-log. Now, we'll use ordinary kriging, which is rather similar. Different to the simple kriging, we use here range and sill, i.e. the variogram function directly instead of the covariance function. 

We provided the function `ordinary_kriging` for you to use. Some points to this function:
* It uses `r` as range and `s` as sill from your `curve_fit`.
* The input `a` is a distance matrix as you created for the variogram cloud (so you can use that!)
* `b` is the distance vector between our data and the target point.
* Many steps in this function are similar to simple kriging, but there is a section called "weight correction", which is a bit more advanced. It ensures that all weights are positive. If interested, here's the reference:_Clayton V. Deutsch, Correcting for negative weights in ordinary kriging, Computers and Geosciences, Volume 22, Issue 7, 1996, Pages 765-773_

````python
def ordinary_kriging(a, b, prop, r, s):
    """Kriging for single location
    
    **Arguments**:
        - a (np.array) = distance between input data points (n x n)
        - b (np.array) = distance vetor between target and input data points (n)
        - prop (np.array) = values of input data (n)
        - r (float) = range parameter of variogram model
        - s (float) = sill parameter of variogram model
        
    **Returns**:
        - result (float) = calculated kriging result
        - pred_var (float) = calculated kriging variance
        - w (np.array) = calculated kriging weights
    """
    #Creating empty matrices
    shape = len(a)
    C = np.zeros((shape + 1, shape + 1))
    c = np.zeros((shape + 1))
    w = np.zeros((shape + 1))

    #filling Matrices based distance matrices and model for spatial correlation
    C[:shape, :shape] = exponential_variogram(a, r, s)
    c[:shape] = exponential_variogram(b, r, s)
    
    np.fill_diagonal(C,0)

    #Fill additional matrix entries, specifically for OK
    C[shape, :] = 1.0
    C[:, shape] = 1.0
    C[shape, shape] = 0.0
    c[shape] = 1.0

    #Solve Kriging equations for weights
    w = np.linalg.solve(C,c)

    #Section: weight correction following Deutsch (1996)
    # 1. Negative Weights to zero
    w[w<0] = 0

    # 2. correct screening effect
    average_magnitude_of_negative_weights = np.mean(w[w>0])
    average_covariance_of_negative_weights = np.mean(c[w>0])
    mask = (w>0) & (c<average_covariance_of_negative_weights) & (w<average_magnitude_of_negative_weights)
    w[mask]=0

    # 3. restandardize to one
    w = w/np.sum(w)
    #End Section: weight correction following Deutsch (1996)

    #Calculating estimate abd variance for kriging
    pred_var = w[shape] + np.sum(w[:shape] * c[:shape])
    result = np.dot(prop, w[:shape])

    return result, pred_var, w
````

In order to produce interpolation results with a variance that resembles more closely the true variance, one maybe wishes to add an additional random component to the kriging results from above.
